# 최종

```python
import requests
import json
import pandas as pd

city = [
"Santa Barbara",
"Ventura",
"Los Angeles",
"San Bernardino",
"Riverside",
"Orange",
"San Diego",
"San Francisco",
"Sacramento",
"San Jose",
"Stockton",
"Santa Rosa",
"Chico",
"Seattle",
"Salt Lake City",
"Denver",
"Las Vegas",
"Albuquerque",
"Phoenix",
"Jamestown",
"Minneapolis",
"Omaha",
"Kansas City",
"Oklahoma City",
"Dallas",
"Manhattan",
"Boston",
"Chicago",
"Washington",
"Atlanta",
"Miami"
]

key = "fac9280fe884d48ac1a9cf7f74fad502"

weather_data = []

for name in city:
  api = f"http://api.openweathermap.org/data/2.5/forecast?q={name}&appid={key}"
  res = requests.get(api)
  res = json.loads(res.text)

  for i in range(res['cnt']):
    forecast = [res['city']['name'],
                res['list'][i]['dt_txt'],
                res['list'][i]['main']['temp_min'], #최저기온, K
                res['list'][i]['main']['temp_max'], #최고기온, K
                res['list'][i]['main']['sea_level'], #해수면 기압, hPa
                res['list'][i]['main']['humidity'], #습도
                res['list'][i]['weather'][0]['main'], #날씨상태
                res['list'][i]['clouds']['all'], #흐림, %
                res['list'][i]['wind']['speed'], #풍속, meter/sec
                res['list'][i]['wind']['deg'], #풍향, degree
                res['city']['sunrise'], #일출시간,유닉스,UTC
                res['city']['sunset'] #일몰시간,유닉스,UTC
                ]
    weather_data.append(forecast)

df = pd.DataFrame(weather_data, columns=['City','Date','Temp_min','Temp_max','Sea_level','Humidity','Weather','Clouds%','Wind_speed','Wind_deg','Sunrise_time','Sunset_time'])
df.to_csv('weather_forecast(3h for 5days).csv', index=False)
df
```
